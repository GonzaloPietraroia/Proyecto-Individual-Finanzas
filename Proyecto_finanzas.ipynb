### Librerias a utilizar

```python
## Librerias para manipular el dato
import pandas as pd
import numpy as np
from pandas_datareader import data as pdr

## Libreria para obtener datos de finanzas
import yfinance as yf

## Libreria para graficar
import plotly.express as px
```

### Cargado de la informacion y transformacion a DataFrame

```python

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
df_url = pd.read_html(url)
df_url = df_url[0]  
for i in ['BF.B','BRK.B','CEG']:                     ##  Borramos 3 empresas que no se encuentran en los datos y generan error
    df_url.drop(df_url[ df_url['Symbol'] == i ].index,inplace=True)
```

```python
yf.pdr_override()
df_empresas = pdr.get_data_yahoo(df_url['Symbol'].tolist(), start="2000-01-01", end="2021-12-31")            ## Informacion sobre las empresas del 2000 al 2021
```

### Retorno GAP y Retorno intradiario

- Cual es el mejor dÃ­a para invertir teniendo en cuenta el retorno de los movimientos intradiarios

- Cual es el mejor dÃ­a para invertir teniendo en cuenta el retorno de los movimiento gap

```python
## Calculamos los retornos gaps y intra con los precios de apertura y cierre de las empresas
cierres = df_empresas["Close"]
aperturas = df_empresas["Open"]
retornos_gaps = np.log(aperturas/cierres.shift(1)).fillna(0)
retornos_intra = np.log(cierres/aperturas).fillna(0)

## Creo dos dataframe vacios para guardar la informacion de los retornos
gap = pd.DataFrame()
intra = pd.DataFrame()

## Les agrego el nombre del dia a los retornos para poder manipular con mas facilidad los datos
retornos_intra['Dia'] = retornos_intra.index.day_name()
retornos_gaps['Dia'] = retornos_gaps.index.day_name()
```

```python
## Lleno el dataframe vacio de gap con los valores importantes para mi grafico
gap['Dia'] = retornos_gaps.groupby('Dia').mean().sum(axis=1).index
gap['Promedio'] =retornos_gaps.groupby('Dia').mean().sum(axis=1).values
gap['Tipo'] = 'GAP'
```

```python
## Lleno el dataframe vacio de intra con los valores importantes para mi grafico
intra['Dia'] = retornos_intra.groupby('Dia').mean().sum(axis=1).index
intra['Promedio'] =retornos_intra.groupby('Dia').mean().sum(axis=1).values
intra['Tipo'] = 'Intra'
```

```python
## Unifico los dos dataframe de gap e intra
intra_gap = pd.concat([gap, intra])
```

```python
## Ordeno los valores por dia
orden_dias = ['Monday','Tuesday','Wednesday','Thursday','Friday']
intra_gap['Dia'] = pd.Categorical(intra_gap['Dia'],orden_dias)
intra_gap.sort_values('Dia' , inplace=True)
```

```python
## DiseÃ±o el grafico
fig = px.bar(intra_gap, x="Dia", y="Promedio", color="Tipo", barmode="group",title='Retorno GAP e Intradiario',width=1000,height=400)
fig.update_layout(
    margin=dict(l=20, r=20, t=60, b=40)
)

fig.show()
```

Podemos determinar viendo el grafico que en base al retorno gap el mejor dia para invertir es el Martes, y en base al retorno intradiario el mejor dia es el Jueves.

### Mejores industrias para invertir

- Cuales son las mejores industrias que pertenecen al SP500 en las cuales se puede invertir

```python
## Creo un dataframe vacio para guardar los con valor y manipular despues

empresa_crecimiento = pd.DataFrame()
crecimiento_lista = []

## Calculo el crecimiento de cada empresa en base al primer registro en el mercado y el ultimo registro

df_empresas["Adj Close"]
for n in df_empresas["Adj Close"]:
    emp = df_empresas["Adj Close"][n]
  
    crecimiento = (emp[-1]*100)/emp[emp.notna()][0]
    crecimiento_lista.append(crecimiento)
    
## Agrego la informacion necesaria a mi dataframe creado    
    
empresa_crecimiento['Empresa'] = df_empresas["Adj Close"].columns.values
empresa_crecimiento['Crecimiento'] = crecimiento_lista
```

```python
## Unifico el dataframe nuevo con la industria de cada empresa

empresa_crecimiento = pd.merge(df_url[['Symbol','GICS Sector']], empresa_crecimiento, left_on='Symbol', right_on='Empresa').drop(columns='Symbol')
```

```python
empresa_crecimiento
```

```python
## Agrupo por industria y calculo el promedio para saber las mejores industrias

top_industrias = empresa_crecimiento.groupby('GICS Sector').mean().sort_values(by='Crecimiento',ascending=False)
```

```python
top_industrias
```

```python
## DiseÃ±o el grafico

fig = px.funnel(top_industrias, x=top_industrias["Crecimiento"].astype(int), y=top_industrias.index,width=1000,title='Industrias con mayor crecimiento')
fig.update_layout(
    yaxis={'title':'Industrias'},
    margin=dict(l=20, r=20, t=75, b=20)
)
fig.show()
```

En el grafico podemos observar cuales fueron las industrias mas recomendadas en base a su crecimiento historico.

### Mayor volatilidad historica

- Cuales fueron los momentos de alta volatilidad que afectaron al SP500

```python
## Calculo las variaciones y la volatilidad

variaciones = df_empresas["Adj Close"].pct_change()
volatilidad = variaciones.rolling(250).std()*100*(250)**0.5
```

```python
## Sumo la volatilidad de todas las empresas para obtener una general por dia

volatilidad = volatilidad.sum(axis=1).sort_index()
```

```python
## Ignoro los valores que no tienen datos

volatilidad = volatilidad[volatilidad.index>'2000-12-31']
```

```python
## DiseÃ±o el grafico

fig = px.area(volatilidad, x=volatilidad.index, y=volatilidad.values,title='Registro de Volatilidad Historica',width=800,height=500)
fig.update_xaxes(            
                 rangeslider_visible=True,
                rangeselector=dict(
                buttons=list([
                dict(count=4, label="4y", step="year", stepmode="backward"),
                dict(count=8, label="8y", step="year", stepmode="backward"),
                dict(count=12, label="12y", step="year", stepmode="backward"),
                dict(count=16, label="16y", step="year", stepmode="backward"),
                dict(step="all")
        ])
    ))



fig.update_layout(
    yaxis={"title": "Volatilidad"},
    margin=dict(l=20, r=20, t=75, b=40)
)
fig.update_traces(line_color='rgb(102,194,165)')

fig.show()
```

En el grafico podemos ver que los momentos con mayor volatilidad fueron en el 2009 y en el 2020
